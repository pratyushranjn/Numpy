## Advance operation with examples


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurantId, 2021, 2022, 2023, 2024]
sales_data = np.array(
    [
        [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
        [2, 120000, 140000, 160000, 190000],  # Beijing Bites
        [3, 200000, 230000, 260000, 300000],  # Pizza Hub
        [4, 180000, 210000, 240000, 270000],  # Burger Point
        [5, 160000, 185000, 205000, 230000],  # Chai Point
    ]
)

print("==== Zomato sales analysis ==== ")
print("\n Sales data shape", sales_data.shape)
print("First 3 restaurant data:\n", sales_data[0:3])
```

```python
# Total sales per year (exclude first column)
# yearly_total = np.sum(sales_data[:, 1:], axis=0)
# print("\nTotal Sales Per Year (2021-2024):")
# print("2021 â†’", yearly_total[0])
# print("2022 â†’", yearly_total[1])
# print("2023 â†’", yearly_total[2])
# print("2024 â†’", yearly_total[3])


# Minimum sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis=1)  # X axis
print(f"\nOverall Minimum Sale: {min_sales}")

# ðŸ”¼ Maximum sales per year
max_sale = np.max(sales_data[:, 1:], axis=0)
print("Maximum Sales Per Year (2021â€“2024):", max_sale)

# ðŸ“Š Average sales per restaurant
average_sales = np.mean(sales_data[:, 1:], axis=1)
print("Average Sales Per Restaurant:", average_sales)
```

```python
cumulative_sum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumulative_sum)

plt.figure(figsize=(8, 6))
plt.plot(np.mean(cumulative_sum, axis=0))
plt.title("Average cumulative sales across all restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector addititon ", vector1 + vector2)

print("\nMultiplication vector ", vector1 * vector2)
print("\nDot Product:", np.dot(vector1, vector2))

angle = np.arccos(
    np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))
)

print(angle)
```

```python
restaurant_types = np.array(['biryani', 'chai', 'pizza', 'samosa'])

# vectorized operation
vectorized_upper = np.vectorize(str.upper)

print('Vectorized Upper:\n', vectorized_upper(restaurant_types))
```
